```python
from bs4 import BeautifulSoup
import requests
news_links = {'politics':[],'kinmel':[],'opinion':[],'nepali-brand':[],'social':[],'art':[],'blog':[],'sports':[],'literature':[],'global':[],'ghumphir':[]}
categories = list(news_links.keys())
for i in categories:
    url = BeautifulSoup(f'https://www.setopati.com/{i}', 'html.parser')
    soup = requests.get(url)
    raw_code = BeautifulSoup(soup.text, "lxml")
    section_code = raw_code.find("section", class_="section special-news cat-list")
    for link in section_code.find_all('a'):
        if "page" not in str(link):
            news_links[i].append(link.get('href'))
        
news_links
```

```python
from bs4 import BeautifulSoup
import requests
n = int(input("Enter how many page you want to scrap = "))
data = {'category':[],'link':[],'news':[]}
categories =  ['politics','kinmel','opinion','nepali-brand','social','art','blog','sports','literature','global','ghumphir']
for i in categories:
    for j in range(1,n+1):
        url = BeautifulSoup(f'https://www.setopati.com/{i}?page={j}', 'html.parser')
        soup = requests.get(url)
        raw_code = BeautifulSoup(soup.text, "lxml")
        section_code = raw_code.find("section", class_="section special-news cat-list")
        for link in section_code.find_all('a'):
            if "page" not in str(link):
                data['link'].append(link.get('href'))
                data['category'].append(i)
        
```

```python
from bs4 import BeautifulSoup
import requests
for i in data['link']:
  url = BeautifulSoup(i, 'html.parser')
  soup = requests.get(url)
  raw_code = BeautifulSoup(soup.text, "lxml")
  div_code = raw_code.find("div", class_="editor-box")
  news = (div_code.text).replace('\n','')
  data['news'].append(news)

import pandas as pd
df = pd.DataFrame(data)
df
```

```python
df.to_csv('nepali_setopati_dataset1.csv')
```
